```python
"""
Example using Dask with the Titanic dataset.
"""
import dask.dataframe as dd
from pysuricata import generate_report

from IPython.display import HTML


TITANIC_URL = (
    "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
)
```

```python
titanic_ddf = dd.read_csv(TITANIC_URL)
html_report = generate_report(titanic_ddf, output_file="titanic_dask_report.html")
print("Dask report saved as 'titanic_dask_report.html'")
HTML(html_report)
```
